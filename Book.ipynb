```python
import networkx as nx
import csv
import matplotlib.pyplot as plt
```

```python
class Book:
    def __init__(self, name, author, userrating, genre):
        self.name = name
        self.author = author
        self.userrating = userrating
        self.genre = genre
```

```python
Books = []
with open("bestsellers with categories.csv",encoding='gb18030',errors='ignore') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        name = row['Name']
        author = row['Author']
        userrating = row['User Rating']
        genre = row['Genre']
        Books.append(Book(name, author, userrating, genre))
```

```python
print(Books[3].genre)
```

```python
BookGraph = nx.Graph()
```

```python
bookSubset = Books[0:1000]
```

```python
for f in bookSubset:
    BookGraph.add_node(f)
```

```python
for f1 in bookSubset:
    for f2 in bookSubset:
        if not f1 == f2:
            if f1.userrating == f2.userrating:
                BookGraph.add_edge(f1,f2)
         
```

```python
options = {'node_color':'black', 'node_size':10, 'width':1}
plt.figure(1, figsize = (30, 30))
nx.draw(BookGraph, **options)
```
