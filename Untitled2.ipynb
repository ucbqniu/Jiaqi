```python
import networkx as nx
import csv
import matplotlib.pyplot as plt

g = nx.DiGraph()
g.add_edges_from([(1, 2), (1, 3), (2, 4), (3, 5), (4, 1), (4, 5)]) 
print(list(nx.all_shortest_paths(g, 1, 4)))
```

```python
options = {'node_color':'red', 'edge_color':'grey', 'node_size':10, 'width':1}
plt.figure(1, figsize = (20, 20))
nx.draw(g, **options)
```
